```python
# libraries
import numpy as np
import matplotlib.pylab as plt
import ruptures as rpt
```

```python
# import data
data = open("test641data.txt").readlines() #loads data into a list
points = np.array( data ).astype(np.float) #converts data to numpy array as floats
```

```python
# change point detection
algo = rpt.Dynp(model="l1", min_size=1, jump=1).fit(points) 
#Dynp= dymanic programming
#model options are "l1", "l2", and "rbf"
#min size= minimum distance between change points 
#jump=subsample(one every "jump" points)
my_bkps = algo.predict(n_bkps=1) #number of breakpoints found
```

```python
# show results
rpt.display(points, my_bkps, figsize= (10,6))
plt.show()
```
